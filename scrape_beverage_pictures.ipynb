### Set Up Bing Search Image API

Reference: https://pypi.org/project/bing-image-downloader/

Before running this script, be sure to add file permission for your terminal if you are using MAC OS. Ref: https://i.stack.imgur.com/YSZ67.png

Note: Google image downloader causes headache on image downloading over the size of 100. Since we need more than 100 pictures per drink category, we now use bing image downloader instead.

```python
import os
import shutil
from bing_image_downloader import downloader
```

```python
os.getcwd()
```

```python
search_list = [
    "7up can", 
    "7up plastic bottle"
]
```

```python
def download_beverage_pictures(beverage_list, amount=1000):
    '''
        Download beverage pictures using Bing Image Downloader.
        Store them in the DATA path
    '''
    DATA = os.getcwd() + "\\7up"
    
    # Remove old DATA dir using shutil, os.remove has permission issues
    if os.path.isdir(DATA):
        shutil.rmtree(DATA)
        
    # Start downloading. May take quite a while.
    for drink in beverage_list:
        downloader.download(drink, 
                        limit=amount,  
                        output_dir=DATA, 
                        adult_filter_off=False, 
                        force_replace=False, 
                        timeout=60, 
                        verbose=True)
    print("Download Finished")
    return
```

```python
download_beverage_pictures(search_list, amount=1000)
```
